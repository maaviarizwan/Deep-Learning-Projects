```python
def initialize_parameters(layers_dims):
    np.random.seed(3)
    parameters = {}
    L = len(layers_dims)            # number of layers in the network

    for l in range(1, L):
        parameters['W' + str(l)] = np.ones((layers_dims[l-1], layers_dims[l])) * 0.1
        parameters['b' + str(l)] = np.zeros((1, layers_dims[l]))
    return parameters
```
