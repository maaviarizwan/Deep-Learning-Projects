```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa','profile_score','lpa'])
```

```python
df
```

```python
def initialize_parameters(layers_dims):
    np.random.seed(3)
    parameters = {}
    L = len(layers_dims)            # number of layers in the network

    for l in range(1, L):
        parameters['W' + str(l)] = np.ones((layers_dims[l], layers_dims[l-1])) * 0.1
        parameters['b' + str(l)] = np.zeros((layers_dims[l], 1))
    return parameters
```

```python
initialize_parameters([2, 2, 1])
```

```python
def linear_forward(A_prev, W, b):
    Z = np.dot(W, A_prev) + b
    return Z
```

```python
def L_layer_forward(X, parameters):
    A = X
    L = len(parameters) // 2 

    for l in range(1, L+1):
        A_prev = A
        Wl = parameters['W'+ str(l)]
        bl = parameters['b'+ str(l)]
        # print("A"+str(l-1)+": ",A_prev)
        # print("W"+str(l)+": ",Wl)
        # print("b"+str(l)+": ",bl)
        # print("--"*20)

        A = linear_forward(A_prev, Wl, bl)
        # print("A"+str(l)+": ", A)

    return A, A_prev
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]

parameters = initialize_parameters([2,2,1])

y_hat,A1 = L_layer_forward(X, parameters)
```

```python
y_hat = y_hat[0][0]
```

```python
def update_parameters(parameters, y, y_hat,A1,X):
    error = float(0.001 * 2 * (y - y_hat))
    
    parameters['W2'][0][0] = parameters['W2'][0][0] + (error * A1[0][0]).item()
    parameters['W2'][0][1] = parameters['W2'][0][1] + (error * A1[1][0]).item()
    parameters['b2'][0][0] = parameters['b2'][0][0] + error
    
    parameters['W1'][0][0] = parameters['W1'][0][0] + (error * parameters['W2'][0][0] * X[0][0]).item()
    parameters['W1'][1][0] = parameters['W1'][1][0] + (error * parameters['W2'][0][0] * X[1][0]).item()
    parameters['b1'][0][0] = parameters['b1'][0][0] + (error * parameters['W2'][0][0]).item()
    
    parameters['W1'][0][1] = parameters['W1'][0][1] + (error * parameters['W2'][0][1] * X[0][0]).item()
    parameters['W1'][1][1] = parameters['W1'][1][1] + (error * parameters['W2'][0][1] * X[1][0]).item()      
    parameters['b1'][1][0] = parameters['b1'][1][0] + (error * parameters['W2'][0][1]).item()
```

```python
update_parameters(parameters, y, y_hat,A1,X)
```

```python
parameters
```

```python
X = df[['cgpa', 'profile_score']].values[2].reshape(2,1)
y = df[['lpa']].values[2][0]

y_hat,A1 = L_layer_forward(X, parameters)
```

```python
# epochs implemenation

parameters = initialize_parameters([2,2,1])
epochs = 5

for i in range(epochs):
    loss = []

    for j in range(df.shape[0]):
        X = df[['cgpa', 'profile_score']].values[j].reshape(2,1)
        y = df[['lpa']].values[j][0]

        y_hat,A1 = L_layer_forward(X, parameters)
        loss.append((y - y_hat[0][0])**2)

        update_parameters(parameters, y, y_hat,A1,X)

        loss.append((y - y_hat[0][0])**2)

    print('Epoch: ', i+1, ' Loss: ', np.array(loss).mean())
parameters
```
